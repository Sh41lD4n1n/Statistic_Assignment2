# Task 1

In a research program on human health risk from recreational contact with water contaminated with pathogenic microbiological material, the National Institute of Water and Atmosphere (NIWA) instituted a study to determine the quality of NZ stream water at a variety of catchment types. This study is documented in McBride et al. (2002) where n = 116 one-liter water samples from sites identified as having a heavy environmental impact from birds (seagulls) and waterfowl. Out of these samples, x = 17 samples contained Giardia cysts. Let θ denote the true probability that a one-liter water sample from this type of site contains Giardia cysts.

## 1. What is the conditional distribution of X, the number of samples containing Giardia cysts, given θ?

Conditional distribution of $X$ is **binomial**. As binomial distribution descrive number of successes (in ourcase failes) $x = 17$ in a sequence of $n = 116$ independent experiment with 2 outcomes with probabilitices $\theta$ and $1 -\theta$.

## 2. Before the experiment, the NIWA scientists elicited that the expected value of θ is 0.2 with a standard deviation of 0.16. Determine the parameters α and β of a Beta prior distribution for θ with this prior mean and standard deviation. (Round α and β to the nearest integer).

We need parameters of Beta distribution of $\theta$:\
We know that:

$$E(\theta) = \frac{\alpha}{\alpha+\beta}$$

$$Var(\theta) = \frac{\alpha \beta}{(\alpha+\beta+1)(\alpha+\beta)^2}$$

$$E(\theta) = 0.2\ \ (1)$$ 
$$Var(\theta) = 0.16^2\ \ (2)$$

Let's solve system of equations: \
Transform equation (1):
$$E(\theta) = \frac{\alpha}{\alpha+\beta} = 0.2$$
$$\alpha = (\alpha+\beta) * 0.2  => \alpha = (\alpha+\beta) * 0.2$$
$$4 \alpha = \beta$$

Use optained equation: $4 \alpha = \beta$ and transform equation (2):\
$$\frac{\alpha \beta}{(\alpha+\beta+1)(\alpha+\beta)^2} = 0.16^2$$
$$\frac{4\alpha^2}{(5 \alpha+1)25 \alpha^2} = 0.0256$$
$$\frac{4}{(5 \alpha+1)25} = 0.0256$$
$$\frac{4*10^4}{25*16*16} = (5 \alpha+1)$$
$$\frac{25}{4} = (5 \alpha+1)$$
$$\alpha = 21/20 = 1.05$$

Apply $4 \alpha = \beta$:
$$\beta = 4 \alpha = 4.2$$

$$\alpha = 1 ; \beta = 4$$

Thus $\theta = Beta(1,4)$

## 3. Find the posterior distribution of θ and summarize it by its posterior mean and standard deviation.

From previous exersizes we know that:
$$P(\theta) = Bin(116,17,\theta)$$
$$P(X|\theta) = Beta(1,4)$$
$$P(X) = \int P(X|\theta)P(\theta)d\theta$$

We need to compute $P(\theta|X)$:
$$P(\theta|X) = \frac{P(\theta)P(X|\theta)}{P(X)}$$

We may simplify calculation by Conjugate prior property: \
$$P(\theta|X) = Beta(1+17,4 + 116 - 17) = Beta(18,103)$$
$$mean = \frac{\alpha}{\alpha+\beta} = 0.14876$$
$$var = \frac{\alpha \beta}{(\alpha+\beta+1)(\alpha+\beta)^2} = 0.00103$$

# 4. Plot the prior, posterior and normalized likelihood.

```python
import numpy as np
from scipy.stats import beta, binom
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0, 1, 100)

likelihood = binom.pmf(17, 116, theta)
prior = beta(1, 4).pdf(theta)
posterior = beta(18, 103).pdf(theta)
```

```python
plt.plot(theta,likelihood,label = "likelihood: P(X|theta)")
plt.plot(theta,prior,label = "prior: P(theta)")
plt.plot(theta,posterior,label = "posterior: P(theta|X)")
plt.xlabel('theta (prob)')
plt.ylabel("function")
plt.legend()
plt.show()
```

## 5. Find the posterior probability that θ < 0.1.

```python
beta(18, 103).cdf(0.1)
```

```python
beta(18, 103).interval(0.95)
```
