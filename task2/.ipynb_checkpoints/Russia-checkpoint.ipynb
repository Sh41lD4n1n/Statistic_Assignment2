```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cord_list = []
for item in pd.read_json("russia.json")['geojson'][0]['coordinates']:
    for cord in item[0]:
        cord_list.append(cord)
```

```python
plt.plot(np.array(cord_list)[:,0],np.array(cord_list)[:,1],'.')
```
